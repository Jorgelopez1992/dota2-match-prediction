```python
import requests 
import json 
response=requests.get('https://api.opendota.com/api/matches/6425768476')
data=json.loads(response.text)
data
```

```python
temp_list=[] #temporary list to store items 
items=[]     #list of itemlists 
p_match_id=[]
heroes=[]
players=[] #player account id 
win=[]    #bolean 
time=[]   #time in unix 
skill=[]
roaming=[]
p_rank=[]  #player rank, tens indicate rank ones indicate stars 
lane=[]
lane_role=[]
p_denies=[]
isRadiant =[] #whether or not the player is on Radiant
p_kda=[]
p_kpm = []
p_gpm=[]
p_deaths=[]
p_kills=[]
p_item0=[]
p_item1=[]
p_item2=[]
p_item3=[]
p_item4=[]
p_item5=[]
p_bp_0=[]
p_bp_1=[]
p_bp_2=[]
```

```python
for player in data['players']:
    p_match_id.append(player['match_id'])
    players.append(player['account_id'])
    heroes.append(player['hero_id'])
    temp_list=[]
    for item in player['purchase']:
        temp_list.append(item)
    items.append(temp_list)
    win.append(player['win'])
    time.append(data['start_time'])
    skill.append(data['skill'])
    roaming.append(player['is_roaming'])
    p_rank.append(player['rank_tier'])
    lane.append(player['lane'])
    lane_role.append(player['lane_role'])
    p_kda.append(player['kda'])
    p_kpm.append(player['kills_per_min'])
    p_kills.append(player['kills'])
    p_item0.append(player['item_0'])
    p_item1.append(player['item_1'])
    p_item2.append(player['item_2'])
    p_item3.append(player['item_3'])
    p_item4.append(player['item_4'])
    p_item5.append(player['item_5'])
    p_deaths.append(player['deaths'])
    p_denies.append(player['denies'])
    p_gpm.append(player['gold_per_min'])
    p_bp_0.append(player['backpack_0'])
    p_bp_1.append(player['backpack_1'])
    p_bp_2.append(player['backpack_2'])
```
